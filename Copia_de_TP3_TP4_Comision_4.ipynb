##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el número de lanzamientos de los dados.
N = 1000

# Lanzar N dados y almacenar los resultados en un array.
resultados = np.random.randint(1, 7, size=N)

# Contar cuántas veces aparece cada número.
conteo = np.bincount(resultados)

# Calcular la probabilidad de obtener cada número al lanzar los dados.
probabilidades = conteo / N

# Imprimir la probabilidad de obtener un 2, 3 o 5.
print("Probabilidades de obtener 2, 3 o 5:", probabilidades[[2, 3, 5]])

# Graficar un histograma de las probabilidades.
plt.hist(probabilidades, bins=6)
plt.show()
```
